```python
def least_differences(a, b, c):
    """
    Return the smallest difference between any two numbers    
    among a, b and c.    
    >>> least_difference(1, 5, -5)
    4    
    """
    diff1 = abs(a - b)
    diff2 = abs(b - c)
    diff3 = abs(a - c)
    return min(diff1, diff2, diff3)
    
```

```python
def mult_by_five(x):
    return 5 * x

def call(fn, arg):
    """Call fn on arg"""
    return fn(arg)

def squared_call(fn, arg):
    """Call fn on the result of calling fn on arg"""
    return fn(fn(arg))

print(
    call(mult_by_five, 1),
    squared_call(mult_by_five, 1), 
    sep='\n'
)
```

```python
# 任务1：编写函数判断乘客是否成年
def is_adult(age, adult_threshold=18):
    """根据年龄判断是否成年"""
    return age >= adult_threshold

# 测试函数
print(is_adult(22))  # 输出: True
print(is_adult(14))  # 输出: False

# 任务2：统计不同性别的生存率（硬编码简化数据）
# 数据格式：[性别, 是否生存（1=生存，0=未生存）]
passengers = [                                         #passagers==data
    ['male', 0],
    ['female', 1],
    ['female', 1],
    ['male', 0],
    ['female', 1]
]

def calculate_survival_rate(data, gender):
    """计算指定性别的生存率"""
    total = sum(1 for p in data if p[0] == gender)          #每找到一个符合条件的乘客，就生成一个 1    #p是 data 里面的每个元素
    survived = sum(1 for p in data if p[0] == gender and p[1] == 1)
    return survived / total if total > 0 else 0

# 输出结果
print(f"男性生存率: {calculate_survival_rate(passengers, 'male'):.1%}")  
print(f"女性生存率: {calculate_survival_rate(passengers, 'female'):.1%}")
```
